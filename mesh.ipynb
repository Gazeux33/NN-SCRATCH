```python
nx,ny = 100,100
```

```python
x = np.linspace(0,1,nx)
y = np.linspace(0,1,ny)
x.shape,y.shape
```

```python
xv,yv = np.meshgrid(x,y)
```

```python
print(xv)
```

```python
print(yv)
```

```python
xv.shape,yv.shape
```

```python
xfv = np.expand_dims(xv.flatten(),-1)
yfv = np.expand_dims(yv.flatten(),-1)
xfv.shape,yfv.shape
```

```python
def custom_model(x, y):
    condition = (x > 0.5) & (y > 0.5)
    return np.where(condition, 1, 0)
```

```python
result = custom_model(xfv,yfv)
result.shape
```

```python
import matplotlib.pyplot as plt
```

```python
figure = plt.figure(figsize=(10,5))
plt.scatter(xfv,yfv,c=result,cmap="rainbow")
plt.title("model predictions")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show();
```
