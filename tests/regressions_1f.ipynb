```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

from nn.Module import Sequential, Linear
from nn.Loss import MSE
from nn.Optimizer import SGD
from nn.Trainer import Trainer
from nn.Dataloader import Dataloader
```

```python
n_samples = 100
n_features = 1
noise = 3.0
```

```python
X,y = make_regression(n_samples=n_samples, n_features=n_features,noise=noise)
X.shape, y.shape
```

```python
fig = plt.figure(figsize=(12,5))
plt.scatter(X, y,marker="^",alpha=1,label="true")
plt.grid()
plt.legend()
plt.show()
```

```python
reg_model = Sequential(
    Linear(n_features,10),
    Linear(10,10),
    Linear(10,1)
)
```

```python
pred = reg_model(X)
pred.shape
```

```python
fig = plt.figure(figsize=(12,5))
plt.scatter(X, y,marker="^",alpha=1,label="true")
plt.scatter(X, pred,marker="v",alpha=1,label="pred")
plt.grid()
plt.legend()
plt.show()
```

```python
batch_size = 32
learning_rate = 0.001

dataloader = Dataloader(batch_size=batch_size,
                        dataset=X,
                        targets=y)
optimizer = SGD(lr=learning_rate)
loss_fn = MSE()

trainer = Trainer(model=reg_model,
                  loss=loss_fn,
                  optimizer=optimizer)
```

```python
hist = trainer.train(dataloader,5)
```

```python
plt.plot(hist,label="loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show();
```
