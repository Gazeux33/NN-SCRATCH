```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_regression

from nn.Module import Sequential, Linear
from nn.Loss import MSE
from nn.Optimizer import SGD
from nn.Trainer import Trainer
from nn.Dataloader import Dataloader
```

```python
n_samples = 1000
n_features = 2
```

```python
X,y = make_regression(n_samples=n_samples, n_features=n_features)
X.shape, y.shape
```

```python
part_to_show = 30
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(X[:part_to_show,0], X[:part_to_show,1], y[:part_to_show],marker="^",alpha=1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
reg_model = Sequential(
    Linear(n_features,10),
    Linear(10,10),
    Linear(10,1)
)
```

```python
pred = reg_model(X[:part_to_show])
pred.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(X[:part_to_show,0], X[:part_to_show,1], y[:part_to_show],marker=".",alpha=1)
ax.scatter(X[:part_to_show,0], X[:part_to_show,1], pred,marker=".",c="orange",alpha=1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
batch_size = 32
learning_rate = 0.001

dataloader = Dataloader(batch_size=batch_size,
                        dataset=X,
                        targets=y)
optimizer = SGD(lr=learning_rate)
loss_fn = MSE()

trainer = Trainer(model=reg_model,
                  loss=loss_fn,
                  optimizer=optimizer)
```

```python
hist = trainer.train(dataloader,1)
```

```python
plt.plot(hist,label="loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show();
```

```python
pred = reg_model(X[:part_to_show])
pred.shape
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(projection='3d')

ax.scatter(X[:part_to_show, 0], X[:part_to_show, 1], y[:part_to_show], marker=".", alpha=1, s=50)

ax.scatter(X[:part_to_show, 0], X[:part_to_show, 1], pred[:part_to_show], marker=".", c="orange", alpha=0.05, s=10)

ax.view_init(elev=10., azim=30)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
